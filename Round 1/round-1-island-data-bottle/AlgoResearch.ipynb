```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df2 = pd.read_csv('trades_round_1_day_-2.csv', delimiter = ";")

# add 999800 to all timestamps in df1
df1 = pd.read_csv('trades_round_1_day_-1.csv', delimiter = ";")
df1['timestamp'] = df1['timestamp'] + 999800

# add 999600 to all time stampes in df0
df0 = pd.read_csv('trades_round_1_day_0.csv', delimiter = ";")  
df0['timestamp'] = df0['timestamp'] + 999600 + 999800

df = pd.concat([df2, df1, df0], ignore_index=True)
df.head()
```

```python
#plot symbol = 'KELP' price over timestamp
kelp_df = df[df['symbol'] == 'SQUID_INK']

# Plot KELP price over time
plt.figure(figsize=(10, 6))
plt.plot(kelp_df['timestamp'], kelp_df['price'], marker='o', linestyle='-')
plt.title('SQUID INK Price Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()


```

```python
trades = pd.read_csv('trades_round_1_day_-2.csv', delimiter = ";")

# Filter trades for symbol 'SQUID_INK'
squid_trades = trades[trades['symbol'] == 'SQUID_INK']

plt.figure(figsize=(10, 6))
plt.plot(squid_trades['timestamp'], squid_trades['price'], marker='o', linestyle='-')
plt.title('SQUID_INK Trades Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
